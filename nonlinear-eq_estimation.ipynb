From $n$ non-linear equations with $x_{1},..., x_n$ unknown variables a multivariate system of nonlinear equations with 
dimensional compatibility can be expressed: <br>

<div class="alert alert-block alert-info">

$f_1(x_1,x_2,...,x_n) = 0$ <br>
$f_2(x_1,x_2,...,x_n) = 0$ <br>
$\vdots$ <br>
$f_n(x_1,x_2,...,x_n) = 0$ <br>

A common method of approximatation for the solutions of $x_1,x_2,...,x_n$ is [Newthon's](README.md#References) method with the roots of terms ${ }_{1,2}$ of the [Taylor series](README.md#References) expansion:<br>
$f(x) = f(x_n) + f'(x_n)(x-x_n) + \frac{f''(x_n)}{2!}(x-x_n)^2 +\cdots + \frac{f^{n}(x_n)}{n!}(x-x_n)^n$ = 0<br>
$f(x) = f(x_n) + f'(x_n)(x-x_n) = 0$ <br>

Generalizing for $n$ variables, we have: <br>
$x_{n+1} = x_n - \frac{f(x_i)}{f'(x_i)} = x_n - J(x)^{-1} \cdot f(x_i)$ <br>

The [Jacobian](README.md#References) matrix is obtained by taking the first derivative of each function with respect to each variable. <br>

$J(x) = \begin{bmatrix}
\frac{\partial f_1}{\partial x_1} & \frac{\partial f_1}{\partial x_2} & \cdots & \frac{\partial f_1}{\partial x_n} \\
\frac{\partial f_2}{\partial x_1} & \frac{\partial f_2}{\partial x_2} & \cdots & \frac{\partial f_2}{\partial x_n} \\
\vdots & \vdots & \ddots & \vdots \\
\frac{\partial f_n}{\partial x_1} & \frac{\partial f_n}{\partial x_2} & \cdots & \frac{\partial f_n}{\partial x_n} \\
\end{bmatrix}$ 

$\therefore$ the equation is rewritten with $x_n$ and $f(x_i)$ as vectors:<br>

$x_n = \begin{bmatrix} x_1 \\ x_2 \\ \vdots \\ x_n \end{bmatrix}$ and 
$f(x_i) = \begin{bmatrix} f_1(x_1,x_2,...,x_n) \\ f_2(x_1,x_2,...,x_n) \\ \vdots \\ f_n(x_1,x_2,...,x_n) \end{bmatrix}$ <br>

So iteration 1 of a Newton-Raphson method is constructed for a multivariate system of nonlinear equations:<br>

$x_{n+1}$ = $x_n = \begin{bmatrix} x_1 \\ x_2 \\ \vdots \\ x_n \end{bmatrix}$ - 
$\begin{bmatrix} \frac{\partial f_1}{\partial x_1} & \frac{\partial f_1}{\partial x_2} & \cdots & \frac{\partial f_1}{\partial x_n} \\ \frac{\partial f_2}{\partial x_1} & \frac{\partial f_2}{\partial x_2} & \cdots & \frac{\partial f_2}{\partial x_n} \\ \vdots & \vdots & \ddots & \vdots \\
\frac{\partial f_n}{\partial x_1} & \frac{\partial f_n}{\partial x_2} & \cdots & \frac{\partial f_n}{\partial x_n} \\
\end{bmatrix}^{-1}$ $\cdot$ $\begin{bmatrix} f_1(x_1,x_2,...,x_n) \\ f_2(x_1,x_2,...,x_n) \\ \vdots \\ f_n(x_1,x_2,...,x_n) \end{bmatrix}$ 
<br>

When $x_{n+1} \approx x_n$ the solution is found. <br>

<div align="right">

<i>(See [refs.](#README.md#References) for more information)</i> 

</div>

This would be the first iteration of the <b>Newton-Raphson</b> method for a multivariate system of nonlinear equations:<br>
Similar <b>Numerical Methods</b> would need to be made until answer(s) converge to their solution.

##### [Beta](README.md#References) (*$\beta$*) distribution.

1. Parameters of distribution $x \sim \beta(\alpha, \beta).$ <br>
- $f(x)$ $\&$ $F(x)$ with $\alpha$ and $\beta$. <br>
- Probabilities $f(x)$ $\&$ $F(x)$ with $\alpha$ and $\beta$. <br>

if $f(x, a, b)=\frac{\Gamma(a+b) x^{a-1}(1-x)^{b-1}}{\Gamma(a) \Gamma(b)}$ $ <i>(see [refs.](README.md#References))</i>

initial guesses can take values $a,b\in[0,1]$.

$\therefore$ The system of linear equations is expressed as:

$\begin{bmatrix}
\frac{40\beta+49\alpha}{\alpha + \beta}-44\\
\frac{81 \alpha \beta}{(\alpha+\beta)^2(1+\alpha+\beta)}-9\\
\end{bmatrix}$
$\begin{bmatrix}
\alpha\\
\beta\\
\end{bmatrix}$

```python
#Libraries
import os
import numpy as np
import pandas as pd
import scipy.stats as st
from scipy.optimize import fsolve
from scipy.integrate import quad
import matplotlib.pyplot as plt
plt.style.use('dark_background')
%matplotlib inline
```

```python
#Scale for B=range
scale, max, min, loc = 9, 49, 40, 44

def fxi(Beta):
    alpha,beta=Beta
    return (((min*beta + max*alpha)/(alpha + beta)) -loc, ((alpha * beta*(scale)**2)/((alpha + beta)**2*(1+alpha + beta)))-scale) #system
try:
    alpha, beta= fsolve(fxi,(1,1))
except:
    print("No solution found")

alpha, beta
```

Now $f(x)$ $\&$ $F(x)$ of $\beta$ is obtained:

```python
#limits for xi
x=np.arange(40,49,0.01)

#Density functions
y=st.beta.pdf(x,alpha,beta, loc=40, scale=9)
Y=st.beta.cdf(x,alpha,beta, loc=40, scale=9)

#Plot of the density functions
plt.style.use('dark_background')
plt.figure(figsize = (20, 12))
plt.plot(x,y, linestyle='dotted', color='white')
plt.plot(x,Y, color='green')

#Style of the plot
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.title('β $f(x, a, b)$ and $F(x, a, b)$ density functions in $(a,b) \in [0,1]$',fontsize=20)
plt.xlabel('x',fontsize=14)
plt.ylabel('Probability',fontsize=14)
plt.legend(['Probability: f(x, a, b)','Accumulated: F(x, a, b)'],fontsize=14)
#Probabilities
df_beta = pd.DataFrame({'x':x, 'f(x)':st.beta.pdf(x,alpha,beta,loc=40,scale=9), 'F(x)':st.beta.cdf(x,alpha,beta,loc=40,scale=9),})

plt.show()

df_beta
```

$\therefore$ <i>The advantage of $F(x)$ cumulative distribution over $f(x)$ density is that the former $F(x) =  P(X \leq x)$ can be defined for continuous, discrete $\&$ mixed distributions.</i> 

##### *Triangular distribution*
- Plot $\&$ probabilities $f(x)$ and $F(x)$.

The Triangular probability density function is given by:
$f(x) = \begin{cases}
\frac{2(x-a)}{(b-a)(c-a)} & \text{for } a \leq x \leq c, \\
\frac{2(b-x)}{(b-a)(b-c)} & \text{for } c \leq x \leq b, \\
0 & \text{otherwise.}
\end{cases}$
and takes as arguments:
- $a$ = lower limit.
- $b$ = upper limit.
- $c$ = mode.<br>

e.g of its usage:<br>

`st.triang.pdf` $[(x, c, loc=a, scale=b-a]$ 

```python
c = .44
x = np.linspace(40,49,100)
t = st.triang.pdf(x,c,loc=40,scale=9)
T = st.triang.cdf(x,c,loc=40,scale=9)
# graph
plt.style.use('dark_background')
fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (20, 8))

ax1.plot(x,t, color='white')
ax2.plot(x,T, color='green')

fig.suptitle('Triangular f(x, c, a, b) & $F(x, c, a, b)$ functions where $x_i \in [0,1]$',fontsize=20)

ax1.set_xlabel('x',fontsize=14)
ax1.set_ylabel('Probability',fontsize=14)
ax1.legend(['Probability: f(x, c, a, b)','Accumulated: F(x, c)'],fontsize=14)
ax1.grid(color='gray', linestyle='--', linewidth=0.5)

ax2.set_xlabel('x',fontsize=14)
ax2.set_ylabel('Probability',fontsize=14)
ax2.legend(['Probability: f(x, c, a, b)','Accumulated: F(x, c)'],fontsize=14)
ax2.grid(color='gray', linestyle='--', linewidth=0.5)

plt.show()

df2_triang = pd.DataFrame({'x':x, 'f(x)':t, 'F(x)':T})
df2_triang
```

<font color= 'orange'> Raw & Stratified Montecarlo Sampling: <font>

Montecarlo methods are an attractive choice as Machine Learning methods in order to sample normalized density functions. <br>

<div class="alert alert-block alert-info">

if $V$ is the volume of an integral 

$$V = \int_{a}^{b} f(x) dx$$
   
and it follows a uniform prob. density function $f(x)$ $X \sim U(a,b)$ it can be modelled with [`np.random.uniform`](README.md#References).<br>

\begin{equation}
f(x) = \begin{cases}
\frac{1}{b-a} & \text{if } a \leq x \leq b\\
0 & \text{otherwise}
\end{cases}
\end{equation}

<i>Nevertheless if the distribution isn't known or has another density it must be obtained from data or modelled with another method.</i><br><br>

From where we've got by the definition the expectancy:<br>

$$E[X] = \int_{a}^{b} x f(x) dx$$

Then, the mean of $x_{i}$ in $V$ is:

$$F = \frac{V}{N} \sum_{i=0}^{N-1} f\left(x_i\right)$$

if $N = N-1$ the equation can be rewritten from the solution of the integral as:
$$F = \frac{b-a}{N-1} \sum_{i=0}^N f\left(x_i\right)$$

But Montecarlo estimations for $N$ random variables $X \sim U(a,b)$ must include the error  $\xi$.<br>
$\therefore$, the estimator can be expressed with the solution of the integral for all possible errors $\xi \in$ $(0,1]$ $\rightarrow$ $a = b$ :

$$\hat{F}^{N} = \frac{b-a}{N-1} \sum_{i=0}^{N} f\bigg[a + \xi(b-a)\bigg]$$

```python
a, b = 0, 1
f = lambda x: (x-a)**3/(b-a)**3
V = quad(f, a, b)[0] 
N = np.logspace(1,7,7,dtype=int) 
U = list(map(lambda N:np.random.uniform(a,b,N),N)) 

def int_montecarlo(f:'Function to integrate',
                   a:'Lower limit of the integral',
                   b:'Upper limit of the integral',
                   U:'Sample of numbers U~[a,b]'):
    return (b-a)/len(U)*np.sum(f(U))

def stratification(N: 'N strata to create',
                   a:'Lower limit of the integral',
                   b:'Upper limit of the integral'):
    {'''Function that receives and creates N number of strata for the stratification method:
    Divide the population into strata and generate a number of observations in each.
    proportional to its probability in order to ensure tha domain is covered and convergence is accelerated.
    '''}
    u= np.random.uniform(a,b,N) 
    i=np.arange(0,N) 
    v = (u+i)/N 
    return v


def approx_MC_EST(f:'Function to integrate',
                  a:'Lower limit of the integral',
                  b:'Upper limit of the integral',
                  N:'Number of terms'):
    {'''Function that receives a function, the limits of the integral, and the number of terms to be used in the
    Monte Carlo method and the stratification method. It returns the value of the integral and the relative error
    of the Monte Carlo method and the stratification method.
    '''}

    U2 = list(map(lambda N:stratification(N,a,b),N)) #Random numbers for the stratification.
    M = [int_montecarlo(f, a, b, U) for U in U] #MC for u1 generated numbers.
    EST = [int_montecarlo(f, a, b, U) for U in U2] #MC for stratified variables.
    df_cv = pd.DataFrame(index=N,columns=['Crude Monte Carlo', 'Relative MC_Error %', 'Stratified Sampling',
                                        'Relative_error M_est %' ], dtype='float') #DataFrame 
    df_cv.index.name = "Terms" #Name df index.
    df_cv.loc[N,"Crude Monte Carlo"] = M #Raw Monte Carlo method on DataFrame over N terms.
    df_cv.loc[N,"Relative MC_Error %"] = np.abs(df_cv.loc[N,"Crude Monte Carlo"]-V)*100/V #Relative error of MC.
    df_cv.loc[N,"Stratified Sampling"] = EST #Stratification Method
    df_cv.loc[N,"Relative_error M_est %"] = np.abs(df_cv.loc[N,"Stratified Sampling"]-V)*100/V #Relative error of the stratification 
    #Plot
    plt.figure(figsize = (20, 12))
    plt.xscale('log') #Logarithmic scale on the x axis   
    plt.plot(df_cv.index,df_cv['Crude Monte Carlo'],marker='o',linestyle='--')
    plt.plot(df_cv.index,df_cv['Stratified Sampling'],marker='o',linestyle='--')
    # Plot Style
    plt.style.use('dark_background')
    plt.grid(linestyle='--')
    plt.title('Monte Carlo with Raw-Stratified Samplings:',fontsize=18)
    plt.xlabel('x',fontsize=14)
    plt.ylabel('Probability',fontsize=14)
    plt.legend(['Monte Carlo Raw Samplings','Monte Carlo Stratified Samplings'],fontsize=14)
    return df_cv, plt.show()

approx_MC_EST(f, a, b, N)[0]
```

<i>$\therefore$ Montecarlo samplings divided into strata are proven to accelerate convergence speeds, nevertheless, random variables should distribute according to the distribution function $f(x)$.</i>
